```python
variables = pd.DataFrame({'normal': np.random.normal(size=15), 
                       'gamma': np.random.gamma(1, size=15), 
                       'poisson': np.random.poisson(size=15)})
print(variables)

variables.cumsum(0).plot()
```

```python
variables.cumsum(0).plot(subplots=True)
```

```python
variables.cumsum(0).plot(subplots=False)
```

```python
variables.cumsum(0).plot(secondary_y='normal')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(7, 12))
for i,var in enumerate(['normal','gamma','poisson']):
    variables[var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('cumulative sum')
```

```python
#Histograms
plt.style.use('seaborn-white')

data = np.random.randn(10000)
print(data)
```

```python
plt.hist(data)
```

```python
plt.hist(data, bins=5, normed=True, alpha=0.6,
         histtype='stepfilled')
```

```python
plt.hist(data, bins=10, normed=False, alpha=0.5,
         histtype='stepfilled')
```

```python
x = np.random.normal(0, 0.8, 1000)
y = np.random.normal(-2, 1, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.5, normed=True, bins=50)

plt.hist(x, **kwargs)
plt.hist(y, **kwargs)
```

```python
counts, bin_edges = np.histogram(data, bins=5)
print(counts)
```

```python
counts, bin_edges = np.histogram(data, bins=50)
print(counts)
```

```python
#Two-Dimensional Histograms
mean = [7, 2]
cov = [[5, 1], [3, 0.8]]
x, y = np.random.multivariate_normal(mean, cov, 10000)
```

```python
plt.hist2d(x, y, bins=20, cmap='Reds')
cb = plt.colorbar()
```

```python
plt.hist2d(x, y, bins=50, cmap='Blues')
cb = plt.colorbar()
```

```python
#plt.hexbin: Hexagonal binnings
plt.hexbin(x, y, gridsize=20, cmap='Reds')

cb = plt.colorbar()
```

```python
plt.hexbin(x, y, gridsize=30, cmap='Greys')

cb = plt.colorbar()
```
